# 🕸️ TechTrack Competitor Intel: Web Scraping & ETL Pipeline

### Executive Summary
* **Objective:** Automate the extraction of competitor laptop pricing and ratings from an unstructured webpage to inform dynamic pricing strategies.
* **Methodology:** Engineered a custom web scraper using Python and `BeautifulSoup` to parse HTML, extract target data, clean string formats, and load the results into a structured Pandas DataFrame.
* **Key Finding:** The competitor's average laptop price is **$1,236.50**, with models ranging from $249 to $2,499.
* **Recommendation:** Utilize the $1,236.50 threshold as the primary benchmark for pricing TechTrack's mid-tier to premium inventory to ensure market competitiveness.
---

```python
import pandas as pd
from bs4 import BeautifulSoup
```

```python
with open('competitor_laptops.html','r') as file:
    html_data=file.read()
```

```python
soup=BeautifulSoup(html_data, 'html.parser')
```

```python
cards=soup.find_all('div',class_='laptop-card')
print(len(cards))
```

```python
soup
```

```python
models = []
prices = []
ratings = []

for card in cards:
    
    
    model_name = card.find('h2', class_='model').text
    models.append(model_name)
    
    price=card.find('span',class_='price').text
    prices.append(price)
    
    
    rating = card.find('span',class_='rating').text
    ratings.append(rating)


print("Models:", models)
print("Prices:", prices)
print("Ratings:", ratings)
```

```python
df=pd.DataFrame({'Models':models,'Prices':prices,'Ratings':ratings})
df
```

```python
df.info()
```

```python
df['Prices']=df['Prices'].astype(float)
```

```python
df['Ratings']=df['Ratings'].str.replace('Stars','')
df['Ratings']=df['Ratings'].astype(float)
```

```python
df.info()
```

```python
average=df['Prices'].mean()
print('Average is:',average)
```
